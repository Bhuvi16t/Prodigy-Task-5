**Create a bar chart or histogram to visualize the distribution of a categorical or continuous variable, such as the distribution of ages or genders in a population.**



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/population acording to male2.csv',encoding= 'utf-8')
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df1=df.drop(['Country Code','Indicator Code'],axis=1)
df1
```

```python
df1
```

```python
chosencountry = 'United States'


countrydata = df1[df1['Country Name'] == chosencountry]


years = countrydata.columns[2:].astype(int)
population = countrydata.iloc[:, 2:].values.flatten()

plt.figure(figsize=(10, 4))
plt.bar(years, population, color='green')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title(f'Population of males {chosencountry} Over the Years')

plt.show()
```

```python
plt.hist(population,bins=30,color='green',edgecolor='black')
plt.title('Population')
plt.show()
```
