```python
#import data
import pandas as pd
from textblob import TextBlob
import matplotlib.pyplot as plt


data = pd.read_csv('twitter_training.csv')
```

#data preprocessing

```python
col_names = ['ID', 'Entity', 'Sentiment', 'Content']
df = pd.read_csv('/content/twitter_training.csv', names=col_names)
df
```

# remoove null values

```python
df.isnull().sum()
```

```python
df=df.dropna()
df.isnull().sum()
```

# sentiment analysis

```python
sentimentCounts = df['Sentiment'].value_counts()
sentimentCounts
```

#Visualize

```python
# figure
plt.figure(figsize=(6,4))
sentimentCounts.plot(kind='bar', color=['red', 'green', 'yellow', 'blue'])
plt.title('Sentiment Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Count of Tweets')
plt.xticks(rotation=0)
plt.show()
```

## *From above graph we can see the most of the tweets are Negative*

```python
brand_data = df[df['Entity'].str.contains('Microsoft', case=False)]
brand_sentiment_counts = brand_data['Sentiment'].value_counts()
brand_sentiment_counts
```

```python

plt.figure(figsize=(6, 6))
plt.pie(brand_sentiment_counts, labels=brand_sentiment_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Sentiment Distribution for Microsoft')
plt.show()
```
